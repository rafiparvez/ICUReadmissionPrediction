```python
import pandas as pd
import numpy as np
import psycopg2
import matplotlib.pyplot as plt
%matplotlib inline
import random
from collections import Counter
```

```python
#Dataframe containing charts data
df_charts_details = pd.read_csv('../data/df_charts_details_2.csv')
print(df_charts_details.shape)
print(df_charts_details.head())
```

```python
#Dataframe containing readmissons data
df_readmission = pd.read_csv('../data/clean_readm_details.csv')

#Creating index in readmissions dataframe
df_readm_indexed = df_readmission.set_index('icustay_id')
print(df_readmission.shape)
print(df_readm_indexed.head())
```

```python
#Creating multi-index dataframe 
df_charts_details['charttime'] = df_charts_details['charttime'].astype('datetime64[ns]')
df_indexed = df_charts_details
df_indexed=df_indexed.dropna(subset=['valuenum'])
df_indexed=df_indexed.drop(['value'],axis=1)

#most frequently measured 250 items
cntr = Counter(df_indexed['itemid'])
freq_item_ids=[item_id for item_id,_ in cntr.most_common(250)]

df_indexed = df_indexed.loc[df_indexed['itemid'].isin(freq_item_ids)]
df_indexed=df_indexed.groupby('icustay_id').apply(lambda x: x.sort_values('charttime'))
df_indexed=df_indexed.drop(['icustay_id'],axis=1)
df_indexed['charttime']=df_indexed.groupby(level=['icustay_id'])['charttime'].apply(lambda x: x-x.iloc[0])
df_indexed = pd.pivot_table(df_indexed, index=['icustay_id', 'charttime'], columns='itemid', values='valuenum')
df_indexed.head()
```

```python
#Merge charts events and readmission data
df_merged_readm_charts = pd.merge(df_indexed.reset_index(), df_readm_indexed.reset_index(),
                  on=['icustay_id'], how='left').set_index(['icustay_id','charttime'])

df_merged_readm_charts.drop(['subject_id','hadm_id'], axis=1, inplace = True)

df_merged_readm_charts.head()
```

```python
eIdx = df_merged_readm_charts.index.levels[0];
nobs = [df_merged_readm_charts.loc[ix].index.shape[0] for ix in eIdx]
plt.hist(nobs,range=(0,1000))
plt.title("Observations per ICU Stay")
plt.show()
```

```python
eIdx = random.choice(list(df_merged_readm_charts.index.levels[0]))
print(eIdx)
df_merged_readm_charts.loc[eIdx, 220045].plot()
plt.ylabel("Heart rate")
plt.xlabel("Time since first encounter")
plt.show()
```

```python
#Normalize dataframe
normalized_df=(df_merged_readm_charts-df_merged_readm_charts.mean())/df_merged_readm_charts.std()
```

```python
#Forward fill missing values
cols = normalized_df.columns.values
normalized_df[cols] = normalized_df.groupby(level=0)[cols].ffill()

#Zero fill missing values
normalized_df.fillna(value=0, inplace=True)
normalized_df.head()
```

```python
normalized_df.loc[243475, 220045].plot()
plt.ylabel("Heart rate (bpm)")
plt.xlabel("Hours since first encounter")
plt.show()
```
